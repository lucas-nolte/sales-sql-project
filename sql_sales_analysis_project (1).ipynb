### Basic Calculation: Revenue per Transaction

```python
cursor.execute("SELECT id, product, (price * quantity) AS revenue FROM sales")
cursor.fetchall()
```

### Overall Metric: Total Revenue

```python
cursor.execute("SELECT SUM(price * quantity) AS total_revenue FROM sales")
cursor.fetchall()
```

### Product Demand: Highest Quantity Sold

```python
cursor.execute("SELECT product, SUM(quantity) AS total_quantity FROM sales GROUP BY product ORDER BY total_quantity DESC LIMIT 3")
cursor.fetchall()
```

### Product Performance: Top Revenue Generating Products

```python
cursor.execute("SELECT product, SUM(price * quantity) AS total_revenue FROM sales GROUP BY product ORDER BY total_revenue DESC LIMIT 3")
cursor.fetchall()
```

### Store Activity: Number of Transactions per Store

```python
cursor.execute("SELECT store, COUNT(*) FROM sales GROUP BY store")
cursor.fetchall()
```

### Store Performance: Total Revenue by Store

```python
cursor.execute("SELECT store, SUM(price * quantity) AS total_revenue FROM sales GROUP BY store")
cursor.fetchall()
```

### Top Transactions: Highest Value Sales

```python
cursor.execute("SELECT *, (price * quantity) AS total_value FROM sales ORDER BY total_value DESC LIMIT 3")
cursor.fetchall()
```
